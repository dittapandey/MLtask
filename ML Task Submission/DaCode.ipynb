```python
import pandas as pd;
import numpy as np;
```

```python
td=pd.read_csv('task2.csv')
```

```python
col=['ID','seg_type','seg_desc','ans','c','per','bin_is_rel']
td.columns=col
```

```python
td.head()
```

```python
td[31:70]
```

```python
td['ans']=td['ans'].replace("I don't use Tinder",'No')
```

```python
td[31:70]
```

```python
is_tinder=td.ans=='Yes'
tinder=td[is_tinder]
```

```python
tinder
```

```python
x=np.zeros((164,2))
y=np.zeros((164,1))
x[:,0]=tinder.c
x[:,1]=tinder.per
y=tinder.bin_is_rel
x
y
```

```python
 from sklearn.linear_model import LogisticRegression
```

```python
 from sklearn.model_selection import train_test_split
```

```python
import sklearn
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
```

```python
classifier=LogisticRegression()
classifier.fit(X_train,y_train)
```

```python
classifier.predict(X_test)
```

```python
lin_predict=classifier.predict(X_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
KNN = KNeighborsClassifier(n_neighbors=5)
```

```python
KNN.fit(X_train, y_train)
```

```python
KNN.predict(X_test)
```

```python
K_predict=KNN.predict(X_test)
```

```python
sklearn.metrics.accuracy_score(y_test, lin_predict)
```

```python
sklearn.metrics.accuracy_score(y_test, K_predict)
```
